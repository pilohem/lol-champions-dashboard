```python
# WANT TO GROUP CHAMPIONS BY TAG (Mage, Tank, Assassin, etc)

# get champion with tags
champion = 'Ahri'
tags = data['data'][champion]['tags']
print(tags)
champion_tags = {champion: tags}
champion_tags
```

```python
def get_champions_with_tags(data):
    champions_list = list(data['data'].keys())
    champions_tags = {}
    for champion in champions_list:
        tags = data['data'][champion]['tags']
        champions_tags[champion] = tags
    return champions_tags
```

```python
champions_tags = get_champions_with_tags(data)
champions_tags
```

```python
types = ['Assassin', 'Fighter', 'Mage', 'Marksman', 'Support', 'Tank']
```

```python
# want to get all mages
champion_type = 'Assassin'
def get_champions_by_type(champion_type, champions_tags):
    champions_list = list(data['data'].keys())
    champions_type_list = []
    for champion in champions_list:
        if champion_type in champions_tags[champion]:
            champions_type_list.append(champion)
    return {champion_type: champions_type_list}
```

```python
result = get_champions_by_type(champion_type, champions_tags)
```
